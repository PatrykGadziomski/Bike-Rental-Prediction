```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#load planets dataset
planets_df = sns.load_dataset("planets")

#print shape of dataframe
print("Shape of data frame: ",planets_df.shape)
```

```python
#Show data frame
planets_df
```

```python
#show descriptive statistics of this dataset
print(planets_df.describe())

#How many missing values can be expected for "orbital_period" based on the shape information and the descriptive statistics ?
```

```python
#calculate and show correlation matrix for orbital_period, mass and distance
corr = planets_df[["orbital_period", "mass", "distance"]].corr()
corr
```

```python
#draw Heatmap with correlations
corr.style.background_gradient(cmap='coolwarm')
```

```python
#further investigate the correlation between the features in a scatter plot

#orbital period shall be plotted on x axis
#distance shall be plotted on y axis
#year shall be visualized through color 
#mass shall be represented by size ('s=')

scatter = planets_df.plot.scatter(x = 'orbital_period',
                                  y = 'distance',
                                  c = 'year',
                                  s = 'mass',
                                  colormap = 'viridis',
                                  sharex = False,)
```

```python
#Analyze values for planet "mass" in a histogram with 15 bins
hist = planets_df['mass'].hist(bins=15, figsize=(10, 5))

#What is special about the shape of the distribution?
```
