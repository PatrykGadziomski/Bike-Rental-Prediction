```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('BikeRentalDaily_test.csv', 'r', encoding='utf-8') as f_test:
    df_test = pd.read_csv(f_test, delimiter=';')
```

```python
df_test
```

```python
# sorted Dataframe, sorted by instant, so that its in the right chronical order
sorted_df = df_test.sort_values(by=['instant'])
index_df = sorted_df.set_index('instant')


# The entries are not complete, there are some missing rows
x = 0
for i in df_test['instant']:
    if i == 1:
        print(i)
    else:
        x += 1
print(x)
index_df
```

```python
# Visualize the distribution of temperature
sns.histplot(df_test['temp'], kde=True)
plt.title('Distribution of Temperature')
plt.show()
```

```python
# Visualize the relationship between temperature and rentals (cnt)
sns.scatterplot(x='temp', y='cnt', data=df_test)
plt.title('Relationship between Temperature and Rentals')
plt.show()
```

```python
sns.scatterplot(x='hum', y='cnt', data=df_test)
plt.title('Relationship between humility and Rentals')
plt.show()
```

```python
sns.lineplot(data=df_test, x="dteday", y="casual")
```

```python
# Visualize the distribution of weather conditions
sns.countplot(x='weathersit', data=df_test)
plt.title('Distribution of Weather Conditions')
plt.show()
```

```python
sns.countplot(x='weekday', data=df_test)
plt.title('Distribution of weekdays')
plt.show()
```

```python
# Visualize the distribution of rentals in different seasons
sns.boxplot(x='season', y='cnt', data=df_test)
plt.title('Distribution of Rentals by Seasons')
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df_test.corr(numeric_only=True)

# Create a heatmap of the entire correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
